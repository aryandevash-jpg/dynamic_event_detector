```python
import pandas as pd
```

```python
fake_news = pd.read_csv('./disaster.csv')
sentiment = pd.read_csv('./sentiment.csv', header=None)
meme = pd.read_csv('./meme.csv')

# Assign column names to sentiment for easier access, assuming the fourth column is text
sentiment.columns = ['id', 'entity', 'sentiment', 'text']

# Combine all text into one list
all_texts = (
    list(fake_news['text']) +
    list(sentiment['text']) +
    list(meme['headline']) # Use 'headline' column for meme
)

print(f'Total texts: {len(all_texts)}')
```

```python
from sentence_transformers import SentenceTransformer

# Load pre-trained SBERT (generic, understands general English)
model = SentenceTransformer('all-MiniLM-L6-v2')

# Convert all texts to embeddings
# This takes 5-10 minutes for 50,000 texts
print('Generating embeddings...')

# Ensure all text columns are strings and handle potential NaN values
cleaned_all_texts = (
    fake_news['text'].fillna('').astype(str).tolist() +
    sentiment['text'].fillna('').astype(str).tolist() +
    meme['headline'].fillna('').astype(str).tolist()
)

embeddings = model.encode(
    cleaned_all_texts,
    batch_size=64,
    show_progress_bar=True
)

print(f'Embedding shape: {embeddings.shape}')
# Output: Embedding shape: (50000, 384)

# Save so you don't have to regenerate
```

```python
from bertopic import BERTopic
from umap import UMAP
from hdbscan import HDBSCAN

# Configure UMAP (dimension reduction)
umap_model = UMAP(
    n_components=3,      # Reduce to 3D for visualization
    n_neighbors=15,
    min_dist=0.1,
    random_state=42
)

# Configure HDBSCAN (clustering algorithm inside BERTopic)
hdbscan_model = HDBSCAN(
    min_cluster_size=10,  # Minimum 10 posts per cluster
    metric='euclidean',
    prediction_data=True
)

# Build BERTopic model
topic_model = BERTopic(
    umap_model=umap_model,
    hdbscan_model=hdbscan_model,
    nr_topics='auto',    # Auto-detect number of topics
    min_topic_size=5
)

# Fit and transform
print('Running BERTopic clustering...')

topics, probs = topic_model.fit_transform(cleaned_all_texts, embeddings)

# See what was discovered
topic_info = topic_model.get_topic_info()
print(topic_info)
```

## Complete Fixed Pipeline
This section implements the data cleaning and modeling improvements to ensure high-quality keywords and accurate event/meme classification.

```python
import re
import pandas as pd
import numpy as np
import requests
from sentence_transformers import SentenceTransformer, util
from bertopic import BERTopic
from sklearn.feature_extraction.text import CountVectorizer

# ── Step 1: Data Preparation & Cleaning ────────────────────────

def clean_text(text):
    if not isinstance(text, str): return ""
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'@\w+|#\w+', '', text)
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip().lower()
    return text if len(text.split()) >= 5 else ""

# Clean and filter — NO meme data in clustering
# Re-using the loaded dataframes: fake_news, sentiment, meme
cluster_texts_raw = (
    list(fake_news['text'].fillna('').astype(str)) +
    list(sentiment['text'].fillna('').astype(str))
)
cluster_texts = [clean_text(t) for t in cluster_texts_raw]
cluster_texts = [t for t in cluster_texts if t]

print(f"Clustering on {len(cluster_texts)} texts (no memes)")

# ── Step 2: SBERT Embeddings ───────────────────────────────
model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(cluster_texts, batch_size=64, show_progress_bar=True)

# ── Step 3: BERTopic with stopwords ───────────────────────
vectorizer = CountVectorizer(
    stop_words="english",
    min_df=2,
    ngram_range=(1, 2)
)

topic_model = BERTopic(
    vectorizer_model=vectorizer,
    min_topic_size=10,
    nr_topics='auto'
)

topics, probs = topic_model.fit_transform(cluster_texts, embeddings)

print("\nTop 5 topics:")
for i in range(min(5, len(topic_model.get_topics()))):
    print(f"Topic {i}: {topic_model.get_topic(i)[:5]}")

# ── Step 4: Meme Similarity Score ─────────────────────────
meme_texts_clean = [clean_text(t) for t in meme['headline'].fillna('').astype(str) if clean_text(t)]
meme_embeddings  = model.encode(meme_texts_clean, show_progress_bar=True)

def meme_similarity_score(cluster_sample):
    emb = model.encode(cluster_sample[:50])
    sims = util.cos_sim(emb, meme_embeddings)
    return sims.max(dim=1).values.mean().item()

df_fixed = pd.DataFrame({'text': cluster_texts, 'topic': topics})

cluster_scores = []
for topic_id in df_fixed['topic'].unique():
    if topic_id == -1: continue
    texts    = df_fixed[df_fixed['topic'] == topic_id]['text'].tolist()
    keywords = [w for w, s in topic_model.get_topic(topic_id)]
    score    = meme_similarity_score(texts)
    cluster_scores.append({
        'topic_id':    topic_id,
        'keywords':    keywords[:5],
        'meme_score':  round(score, 3),
        'total_posts': len(texts)
    })

scores_df = pd.DataFrame(cluster_scores)

# ── Step 5: GDELT & Final Results ─────────────────────────
def verify_gdelt(keywords):
    clean_kw = [k for k in keywords if len(k) > 3]
    if len(clean_kw) < 2: return 0

    params = {
        'query':      ' '.join(clean_kw[:3]),
        'mode':       'artlist',
        'format':     'json',
        'maxrecords': 50
    }
    try:
        r = requests.get('https://api.gdeltproject.org/api/v2/doc/doc', params=params, timeout=10)
        return len(r.json().get('articles', []))
    except:
        return 0

final_results = []
for _, row in scores_df.iterrows():
    news_count = verify_gdelt(row['keywords'])
    meme_score = row['meme_score']

    if news_count > 50 and meme_score < 0.4: label = 'REAL EVENT'
    elif meme_score > 0.6: label = 'MEME'
    elif news_count > 10 and meme_score < 0.5: label = 'POSSIBLE EVENT'
    else: label = 'NOISE'

    final_results.append({
        'topic_id':  row['topic_id'],
        'keywords':  row['keywords'],
        'news_count': news_count,
        'meme_score': meme_score,
        'label':     label
    })

results_df = pd.DataFrame(final_results)
results_df.to_csv('fixed_approach_results.csv', index=False)
display(results_df.head(20))
```

Fine-Tuned SBERT

```python
from sentence_transformers import InputExample
import pandas as pd

# 1. Load the results from our fixed pipeline
results_df = pd.read_csv('fixed_approach_results.csv')

# 2. Create training examples
train_examples = []

# Using Cluster Labels: Real Events vs Memes
# We iterate through the classified clusters and pair texts
for _, row in results_df.iterrows():
    topic_id = row['topic_id']
    label = row['label']

    # Get texts belonging to this specific topic
    topic_texts = df_fixed[df_fixed['topic'] == topic_id]['text'].tolist()

    if label == 'REAL EVENT':
        # Create pairs of similar event texts (Label 1.0)
        for i in range(min(len(topic_texts)-1, 10)):
            train_examples.append(InputExample(texts=[topic_texts[i], topic_texts[i+1]], label=1.0))
    elif label == 'MEME':
        # Create pairs of similar meme texts (Label 1.0)
        for i in range(min(len(topic_texts)-1, 10)):
            train_examples.append(InputExample(texts=[topic_texts[i], topic_texts[i+1]], label=1.0))

print(f'Created {len(train_examples)} training examples from clusters.')

# 3. Add samples from the original labeled datasets for better grounding
# Disaster (Real) vs Sarcastic (Meme)
real_samples = fake_news[fake_news['target'] == 1]['text'].dropna().sample(500).tolist()
meme_samples = meme['headline'].dropna().sample(500).tolist()

# Add contrastive pairs (Real text vs Meme text -> Label 0.0)
for r, m in zip(real_samples, meme_samples):
    train_examples.append(InputExample(texts=[r, m], label=0.0))

print(f'Total training examples: {len(train_examples)}')
```

```python
from torch.utils.data import DataLoader
from sentence_transformers import losses

# 1. Prepare DataLoader
train_dataloader = DataLoader(train_examples, shuffle=True, batch_size=16)

# 2. Use CosineSimilarityLoss for regression-like similarity
train_loss = losses.CosineSimilarityLoss(model=model)

# 3. Fine-tune the model
print('Starting fine-tuning...')
model.fit(
    train_objectives=[(train_dataloader, train_loss)],
    epochs=1,
    warmup_steps=100,
    show_progress_bar=True
)

# Save the fine-tuned model
model.save('fine_tuned_sbert_v1')
print('Model saved as fine_tuned_sbert_v1')
```
